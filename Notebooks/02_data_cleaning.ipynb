```python
import pandas as pd
import numpy as np
```

```python
spot = pd.read_csv(
    "data/nifty_spot_5min.csv",
    parse_dates=['date'],
    index_col='date'
)

futures = pd.read_csv(
    "data/nifty_futures_5min.csv",
    parse_dates=['date'],
    index_col='date'
)

options = pd.read_csv(
    "data/nifty_options_5min.csv",
    parse_dates=['date'],
    index_col='date'
)
```

```python
spot = spot.sort_index()
futures = futures.sort_index()
options = options.sort_index()

spot = spot[~spot.index.duplicated()]
futures = futures[~futures.index.duplicated()]
```

```python
options = options.reset_index().drop_duplicates().set_index('date')
```

```python
#Trading Hours Filter
```

```python
market_open = "09:15"
market_close = "15:30"

spot = spot.between_time(market_open, market_close)
futures = futures.between_time(market_open, market_close)
options = options.between_time(market_open, market_close)
```

```python
futures['open_interest'] = futures['open_interest'].clip(lower=0)
```

```python
options['open_interest'] = options['open_interest'].clip(lower=0)
```

```python
assert spot.index.equals(futures.index)
assert options.index.nunique() == spot.shape[0]
```

```python
import os
os.makedirs("data/processed", exist_ok=True)
```

```python
spot.to_csv("data/processed/nifty_spot_clean_5min.csv")
futures.to_csv("data/processed/nifty_futures_clean_5min.csv")
options.to_csv("data/processed/nifty_options_clean_5min.csv")
```
